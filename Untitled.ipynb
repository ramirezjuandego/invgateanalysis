Alcance: hacer una caracterización de los datos exportados de la plataforma de uso interno de la oficina


```python
import pandas as pd
import numpy as np
from datetime import datetime

%matplotlib inline  


df = pd.read_csv('invgate.csv', sep =';', parse_dates=['Fecha de cierre', 'Fecha de creación'])

df = df[~df.index.duplicated(keep='first')]


df.head(10)
```

```python
import seaborn as sns



sns.countplot(y=df.Secretaría, data=df, order = df['Secretaría'].value_counts().index)
```

```python
sinsecbool = df['Secretaría'] == 'Sin Secretaría'
sinsec = df[sinsecbool]

sinsec.tail(60)
```

```python
"""agente sin secretaría"""


sns.countplot(y=sinsec.Agente, data=sinsec, order = sinsec['Agente'].value_counts().index)
```

```python
"""sin secretaría y extrayendo el usuario Monitor"""

sinsecbool2 = df['Agente'] != 'Monitor Nagios'
sinsec2 = df[sinsecbool2]

sinsec.tail(60)

```

```python

sns.countplot(y=sinsec2.Agente, data=sinsec2, order = sinsec2['Agente'].value_counts().index)
```
